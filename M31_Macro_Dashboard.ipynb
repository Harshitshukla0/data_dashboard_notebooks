# 2Y and 10Y Yield

```python


# URL of the page to scrape
no_of_pages = 28

yield_2y_10y = pd.DataFrame()

for i in range(14,no_of_pages+1):
  url = f'https://home.treasury.gov/resource-center/data-chart-center/interest-rates/TextView?type=daily_treasury_yield_curve&field_tdr_date_value=all&page={i}'

  # Send a GET request to the page
  response = requests.get(url)
  response.raise_for_status()  # Raise an exception for HTTP errors

  # Parse the HTML content of the page
  soup = BeautifulSoup(response.content, 'html.parser')

  # Find the table containing the data
  table = soup.find('table', {'class': 'views-table'})

  # Extract the header row
  headers = [header.text for header in table.find_all('th')]

  # Extract the data rows
  rows = []
  for row in table.find_all('tr')[1:]:
      cols = row.find_all('td')
      cols = [col.text.strip() for col in cols]
      rows.append(cols)

  # Create a DataFrame from the extracted data
  temp = pd.DataFrame(rows, columns=headers)

  # Display the DataFrame
  # yield_2y_10y.sort_values('Date',ascending=False)
  temp = temp[['Date', '2 Yr', '10 Yr']]
  yield_2y_10y = pd.concat([yield_2y_10y, temp])


yield_2y_10y = yield_2y_10y[yield_2y_10y['Date']>='01/01/2010']
yield_2y_10y.head()
yield_2y_10y_curr = yield_2y_10y[yield_2y_10y['Date']==yield_2y_10y['Date'].max()]
yield_2y_10y_curr.columns = ['date','metric_1','metric_2']
yield_2y_10y_curr['metric'] = '2Y Yield'
yield_2y_curr = yield_2y_10y_curr[['date','metric', 'metric_1']]
yield_2y_curr.columns = ['date','metric','value']
yield_10y_curr = yield_2y_10y_curr[['date','metric_2']]
yield_10y_curr.columns = ['date','value']
yield_10y_curr['metric'] = '10Y Yield'
yield_10y_curr
```
